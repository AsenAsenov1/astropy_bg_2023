```python
data_full = np.loadtxt(fname='data/03D1au.csv', delimiter=',',skiprows=1)
```

```python
mjd = data_full[:,0]
flux_g = data_full[:,1]
flux_r = data_full[:,3]
flux_i = data_full[:,5]
flux_z = data_full[:,7]
index_g = ~np.isnan(data_full[:,1])
index_r = ~np.isnan(data_full[:,3])
index_i = ~np.isnan(data_full[:,5])
index_z = ~np.isnan(data_full[:,7])
```

```python
plt.plot(mjd[index_g], flux_g[index_g], '^-', color='blue')
plt.plot(mjd[index_r], flux_r[index_r], 'x-', color='green')
plt.plot(mjd[index_i], flux_i[index_i], 'p-', color='yellow')
plt.plot(mjd[index_z], flux_z[index_z], '1-', color='red')
```

```python
fig = plt.figure(figsize=[16.0, 4.0])

ax1 = fig.add_subplot(1,4,1)
ax2 = fig.add_subplot(1,4,2,sharey=ax1)
ax3 = fig.add_subplot(1,4,3,sharey=ax1)
ax4 = fig.add_subplot(1,4,4,sharey=ax1)

ax1.plot(mjd[index_g], flux_g[index_g],'o-',color='blue')
ax1.set_xlabel('MJD')
ax1.set_ylabel('Flux')
ax1.set_ylim([-10,1600])

ax2.plot(mjd[index_r], flux_r[index_r],'o-',color='salmon')
ax2.set_xlabel('MJD')
ax2.set_ylabel('Flux')
#ax2.set_ylim([-10,800])

ax3.plot(mjd[index_i], flux_i[index_i],'o-',color='#beaed4')
ax3.set_xlabel('MJD')
ax3.set_ylabel('Flux')
#ax3.set_ylim([-10,800])

ax4.plot(mjd[index_z], flux_z[index_z],'o-',color='#386cb0')
ax4.set_xlabel('MJD')
ax4.set_ylabel('Flux')
#ax4.set_ylim([-10,800])

fig.tight_layout()
fig.savefig('myplot.png')
fig.savefig('myplot.pdf')
fig.savefig('myplot.ps')
```
